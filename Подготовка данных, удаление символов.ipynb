```python
import pandas as pd
import re
```

```python
def clean_text(text):
    if isinstance(text, str):
        text = re.sub(r'[!?()«»“”]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
input_path = '/content/drive/MyDrive/Datasets/dataset_labeled.csv'
df = pd.read_csv(input_path, encoding='utf-8')
```

```python
df['text1'] = df['text1'].apply(lambda x: clean_text(x))
df['text2'] = df['text2'].apply(lambda x: clean_text(x))
```

```python
output_path = '/content/drive/MyDrive/Datasets/dataset_step1_1.csv'
df.to_csv(output_path, index=False, encoding='utf-8')
```
