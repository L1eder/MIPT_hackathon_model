```python
input_path = '/content/drive/MyDrive/Datasets/dataset_processed_step1.csv'
df = pd.read_csv(input_path)
```

```python
df1 = df[['id', 'title', 'status']]
df2 = df[['id', 'description', 'status']]
```

```python
output_path1 = '/content/drive/MyDrive/Datasets/dataset_title.csv'
output_path2 = '/content/drive/MyDrive/Datasets/dataset_description.csv'
```

```python
df1.to_csv(output_path1, index=False, encoding='utf-8')
df2.to_csv(output_path2, index=False, encoding='utf-8')

print(f'''
Данные успешно разделены:
1. Файл с заголовками: {output_path1}
   Колонки: {', '.join(df1.columns)}

2. Файл с описаниями: {output_path2}
   Колонки: {', '.join(df2.columns)}
''')
```

```python
def generate_pairs(df, text_column):
    pairs = []
    ids = df['id'].unique()

    for id1, id2 in itertools.combinations(ids, 2):
        row1 = df[df['id'] == id1].iloc[0]
        row2 = df[df['id'] == id2].iloc[0]

        pairs.append({
            'id1': id1,
            'id2': id2,
            'text1': row1[text_column],
            'text2': row2[text_column],
            'label': 1 if row1['status'] == row2['status'] else 0
        })

    return pd.DataFrame(pairs)
```

```python
title_pairs = generate_pairs(df1, 'title')
desc_pairs = generate_pairs(df2, 'description')
```

```python
final_df = pd.concat([title_pairs, desc_pairs], ignore_index=True)
```

```python
final_df.to_csv('/content/drive/MyDrive/Datasets/dataset_final.csv', index=False, encoding='utf-8')

print(f'''
Создано:
- {len(title_pairs)} пар заголовков
- {len(desc_pairs)} пар описаний
Итоговый датасет: {len(final_df)} строк
''')
```
